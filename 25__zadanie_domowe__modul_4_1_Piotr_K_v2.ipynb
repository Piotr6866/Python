### 2. Pozyskanie i ogólne zrozumienie danych

Zbiór danych (25__iris.csv) zawiera informacje o trzech gatunkach irysów: Iris setosa, Iris versicolor i Iris virginica.

Dane obejmują pomiary czterech cech: długość i szerokość działki kielicha oraz długość i szerokość płatka.
Każdy wiersz w zbiorze danych reprezentuje pojedynczy kwiat, a wartości pomiarów są podane w centymetrach.
Zbiór składa się z 150 próbek, po 50 dla każdego gatunku.

Kolumny:
* **długość kielicha (sepal length)** - Długość kielicha w cm
* **szerokość kielicha (sepal width)** - Szerokość kielicha w cm
* **długość płatka (petal length)** - Długość płatka w cm
* **szerokość płatka (petal width)** - Szerokość płatka w cm
* **klasa (class)** - Klasa irysa (setosa, versicolor, virginica)

### 4. Wstępna eksploracja danych (EDA – Exploratory Data Analysis) 

```python

# Pokaż wyliczenia dla: (średnia, mediana, odchylenie standardowe, min, max, kwartyl 25, kawrtyl 75) 
num_cols = df.select_dtypes(include='number')

stats = pd.DataFrame({
    'min': num_cols.min(),
    'kwartyl_25%': num_cols.quantile(0.25),
    'mediana (kwartyl 50%)': num_cols.median(),
    'kwartyl_75%': num_cols.quantile(0.75),
    'max': num_cols.max(),
    'średnia': num_cols.mean().round(2),
    'odchylenie_std': num_cols.std().round(2)
})

stats
```

```python
# Wykonaj mi wykresy dla powyższych wyliczeń ( OGÓLNE )
num_cols = df.select_dtypes(include='number')

# Obliczenia statystyk
stats = pd.DataFrame({
    'min': num_cols.min(),
    'kwartyl_25%': num_cols.quantile(0.25),
    'mediana': num_cols.median(),
    'kwartyl_75%': num_cols.quantile(0.75),
    'max': num_cols.max(),
    'średnia': num_cols.mean().round(2),
    'odchylenie_std': num_cols.std().round(2)
})

# Wykres: Średnia
plt.figure(figsize=(8, 5))
sns.barplot(x=stats.index, y=stats['średnia'])
plt.title('Średnia dla kolumn')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Wykres: Odchylenie standardowe
plt.figure(figsize=(8, 5))
sns.barplot(x=stats.index, y=stats['odchylenie_std'])
plt.title('Odchylenie standardowe dla kolumn')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Wykres: Boxplot dla kolumn numerycznych
plt.figure(figsize=(8, 5))
sns.boxplot(data=num_cols)
plt.title('Boxplot kolumn numerycznych')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# List of numerical columns
numerical_columns = ['długość kielicha (sepal length)', 'szerokość kielicha (sepal width)','długość płatka (petal length)', 'szerokość płatka (petal width)']
 ##x='klasa (class)', y =['długość kielicha (sepal length)', 'szerokość kielicha (sepal width)','długość płatka (petal length)', 'szerokość płatka (petal width)'])
# Create a plot for each numerical column
for column in numerical_columns: 
    plt.figure(figsize=(10, 6))
    
# Plot histogram with KDE (Kernel Density Estimate) for trend line
    ###sns.histplot(data=df, x=column, hue='class', element='step', stat='density', common_norm=False, kde=True)
    sns.histplot(data=df, x=column, hue='klasa (class)', element='step', stat='density', common_norm=False, kde=True)
    
    # Add title and labels
    plt.title(f'Histogram dla {column} dla poszczególnych Klas Irysów')
    plt.xlabel(column)
    plt.ylabel('Rozmiar w [cm]')
    
    # Save the plot as a PNG file
    ###plt.savefig(f'histogram_{column}.png')
    plt.show()
```

```python
# Wyświetlenie macierzy korelacji dla kolumn numerycznych jako heatmapy. Skala od -1 do +1

# Calculate the correlation matrix for numerical columns
correlation_matrix = df.iloc[:, :-1].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa korelacji dla kolumn numerycznych')

plt.show()
```

```python
# Wykonaj macierze korelacji dla każdej z trzech dostępnych klas irysów

# Get unique classes
classes = df['klasa (class)'].unique()

# Create a plot for each class
for iris_class in classes:
    # Filter the dataframe for the current class
    class_df = df[df['klasa (class)'] == iris_class]
    
    # Calculate the correlation matrix
    corr_matrix = class_df.drop(columns='klasa (class)').corr()
    
    # Plot the heatmap
    plt.figure(figsize=(8, 6))
    ##sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True)
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Macierze korelacji dla: {iris_class}')
    plt.tight_layout()

    plt.show()
```

```python
agg_df = df.groupby('klasa (class)',as_index=False).agg({'długość kielicha (sepal length)': ['mean', 'median'], 'szerokość kielicha (sepal width)': ['mean', 'median'], 'długość płatka (petal length)': ['mean', 'median'], 'szerokość płatka (petal width)': ['mean', 'median']})
```

### 6. Raportowanie i prezentacja wyników

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Wykres pudełkowy dla zbioru irysów podzielonych na Klasy')

parameters = [
    'długość kielicha (sepal length)',
    'szerokość kielicha (sepal width)',
    'długość płatka (petal length)',
    'szerokość płatka (petal width)'
]

for ax, param in zip(axes.flatten(), parameters):
    sns.boxplot(
        x='klasa (class)', 
        y=param, 
        data=df, 
        ax=ax,
        hue='klasa (class)',   # 🔹 jawne przypisanie koloru do klasy
        palette="Set2",        # 🔹 paleta kolorów
        legend=False           # 🔹 brak zbędnej legendy
    )
    ax.set_title(f'Wykres pudełkowy {param.replace("_", " ").title()}')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```
